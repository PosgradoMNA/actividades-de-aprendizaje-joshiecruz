```python
#Probando diferentes numeros de cluster para 20 posibles
kmeans = KMeans(n_init=20)

visualizer = KElbowVisualizer(kmeans, k=(1,20),metric='distortion',timings=False,)
visualizer.fit(gdf[['LATITUD','LONGITUD']])  
visualizer.show()
```

```python
kmeans = KMeans(n_clusters = 4, init ='k-means++')
kmeans.fit(gdf[['LATITUD','LONGITUD']]) 
gdf['cluster_sub'] = kmeans.fit_predict(gdf[['LATITUD','LONGITUD']])
centerssub = kmeans.cluster_centers_
gdf.head()
    
```

```python
X = gdf[['LONGITUD', 'LATITUD']]
kmeans = KMeans(n_clusters=4).fit(X)
centroids = kmeans.cluster_centers_
labels = kmeans.predict(X) 
C = kmeans.cluster_centers_
C_DF = pd.DataFrame(C)
C_DF["Coordinates"] = list(zip(C_DF[0], C_DF[1])) 
C_DF["Coordinates"] = C_DF["Coordinates"].apply(Point) 
puntos_centroides = gpd.GeoDataFrame(C_DF, geometry="Coordinates")
puntos_centroides
  
fig, gax = plt.subplots(figsize=(15,10))
world.query("name == 'Mexico'").plot(ax = gax, edgecolor='black', color='white',cmap=cmap) #filtramos por pais

puntos_centroides.plot(ax=gax, color='dodgerblue', alpha = 1, markersize = 1000, marker ="o") 
#points.plot(ax=gax, alpha = 0.7, c=kmeans.labels_) 
gax.set_xlabel('Longitud')
gax.set_ylabel('Latitud')
gax.set_title('Aguas SUPERFICIALES - '+ "KMEANS")
plt.show()
```

```python
dfCalidadSuperficial.isna().sum().sort_values(ascending=False)
```

```python
#borramos 648 filas que no representan ningun dato
dfCalidadSuperficial.dropna(subset=['LONGITUD'], inplace = True)
dfCalidadSuperficial.CALIDAD_DBO.fillna(value=dfCalidadSuperficial.CALIDAD_DBO.mode()[0], inplace=True)
dfCalidadSuperficial.CALIDAD_DQO.fillna(value=dfCalidadSuperficial.CALIDAD_DQO.mode()[0], inplace=True)
dfCalidadSuperficial.CALIDAD_SST.fillna(value=dfCalidadSuperficial.CALIDAD_SST.mode()[0], inplace=True) 
dfCalidadSuperficial.CALIDAD_COLI_FEC.fillna(value=dfCalidadSuperficial.CALIDAD_COLI_FEC.mode()[0], inplace=True) 
dfCalidadSuperficial.CALIDAD_E_COLI.fillna(value=dfCalidadSuperficial.CALIDAD_E_COLI.mode()[0], inplace=True) 
dfCalidadSuperficial.CALIDAD_ENTEROC.fillna(value=dfCalidadSuperficial.CALIDAD_ENTEROC.mode()[0], inplace=True)
dfCalidadSuperficial.CALIDAD_TOX_D_48.fillna(value=dfCalidadSuperficial.CALIDAD_TOX_D_48.mode()[0], inplace=True)
dfCalidadSuperficial.SEMAFORO.fillna(value=dfCalidadSuperficial.SEMAFORO.mode()[0], inplace=True)
```

```python
dfCalidadSuperficial.info()
```

```python
X= dfCalidadSuperficial[["LATITUD","LONGITUD",'CALIDAD_DBO', 'CALIDAD_DQO','CALIDAD_SST', 'CALIDAD_COLI_FEC','CALIDAD_E_COLI','CALIDAD_ENTEROC','CALIDAD_TOX_D_48']]
y=dfCalidadSuperficial[['SEMAFORO']]
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import learning_curve, validation_curve, cross_validate, RepeatedStratifiedKFold, GridSearchCV, train_test_split, RepeatedKFold
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
lbe = LabelEncoder()
y["SEMAFORO_Type"]= lbe.fit_transform(dfCalidadSuperficial['SEMAFORO'])
y["SEMAFORO_Type"].unique()
X['CALIDAD_DBO']=lbe.fit_transform(X['CALIDAD_DBO'])
X['CALIDAD_DQO']=lbe.fit_transform(X['CALIDAD_DQO'])
X['CALIDAD_SST']=lbe.fit_transform(X['CALIDAD_SST'])
X['CALIDAD_COLI_FEC']=lbe.fit_transform(X['CALIDAD_COLI_FEC'])
X['CALIDAD_E_COLI']=lbe.fit_transform(X['CALIDAD_E_COLI'])
X['CALIDAD_ENTEROC']=lbe.fit_transform(X['CALIDAD_ENTEROC'])
X['CALIDAD_TOX_D_48']=lbe.fit_transform(X['CALIDAD_TOX_D_48'])
```

```python
#Decision Tree
seed = 1
x_train, x_test, y_train, y_test = train_test_split(X, y[['SEMAFORO_Type']], test_size=0.2, random_state=seed)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X.shape
```

```python
y.shape
```

```python
X.head
```

```python
y.head
```

```python
# Crear Arbol de decision con profundidad = 5
decisionTree = tree.DecisionTreeClassifier(criterion='gini',
                                            min_samples_split=20,
                                            min_samples_leaf=5,
                                            max_depth = 5,
                                            class_weight="balanced")
# ejecutamos el arbol
decisionTree.fit(x_train, y_train)
print("::::::::FEATURE IMPORTANCE DECISION TREE::::::::")
for score, name in zip(decisionTree.feature_importances_, X.columns):
    print(round(score, 2), name)
```

```python
plt.barh(X.columns, decisionTree.feature_importances_)
plt.show()
```

```python
tree_clf = DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf.fit(x_train, y_train)
y_pred_tree = tree_clf.predict(x_test)
```

```python
target_names = ['class 1', 'class 2', 'class 3']
print(classification_report(y_test, y_pred_tree, target_names=target_names))
```

```python
cm = confusion_matrix(y_test,y_pred_tree)
displ = ConfusionMatrixDisplay(confusion_matrix=cm)
displ.plot()
plt.show()
```

```python
# Random forest


rnd_clf = RandomForestClassifier(n_estimators=500, random_state=42)
X.data = X
y.target = y[['SEMAFORO_Type']]
rnd_clf.fit(X.data, y.target)
for score, name in zip(rnd_clf.feature_importances_, X.data.columns):
    print(round(score, 2), name)
```

```python
plt.barh(X.columns, rnd_clf.feature_importances_)
plt.show()
```
